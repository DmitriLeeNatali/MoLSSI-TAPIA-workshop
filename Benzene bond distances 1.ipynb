```python
import numpy
```

```python
distance_file = os.path.join('benzene.xyz')
```

```python
distances = numpy.genfromtxt(distance_file, dtype='unicode', skip_header=2)
headers = distances[:,0]
print(headers)
#print(distances)
```

```python
data = distances[:,1:]
print(data)
```

```python
A = data.astype(numpy.float)
print(A)
```

```python
A = distances[:,1:]
print(A)
```

```python
A = A.astype(numpy.float)
print(A)
```

```python
A_1 = numpy.zeros((12,12))
for num1 in range(12):
    for num2 in range(12):
        A_1[num1,num2] = (numpy.sqrt((A[num1,0]-A[num2,0])**2+(A[num1,1]-A[num2,1])**2+(A[num1,2]-A[num2,2])**2))
print(A_1)
```

```python
for i in range(12):
    for j in range(12):
        print('{0:2s} to {1}: {2:.5f}'.format(headers[i], headers[j], A_1[i,j]))
```
